```python
# Standard imports
import os
import sys

# Setting up enviornment for neurite
sys.path.append('../../../')
os.environ['NEURITE_BACKEND'] = 'pytorch'

# OPTIONAL: Importing matplotlib for visualization purposes
import matplotlib.pyplot as plt

# Importing samplers from neurite
from neurite.torch.random import (
    Sampler,
    Uniform,
    Fixed,
    Normal,
    Bernoulli,
    Poisson,
    RandInt,
    LogNormal
)
```

```python
sampler = Uniform(13, 14)
samples = sampler((128 ** 2,))
print(sampler.serialize())

# Plot the histogram
plt.hist(samples.numpy(), bins=128, edgecolor='black', alpha=0.7)
plt.title("Frequency Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
